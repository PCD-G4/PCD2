# Transformasi Log
Import Library

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Path Gambar dan Membaca Gambar

```python
img_path = "fourierspectrum.jpg"
img = cv2.imread(img_path, 0)
print("Image shape:", img.shape)
```

Fungsi Transformasi Logaritma

```python
def log_transformation(r, c):
    r = r.astype(np.float32)
    s = c * np.log(1 + r)
    s = rescale_image(s)
    return s.astype(np.uint8)
```

Rescale Gambar

```python
def rescale_image(img):
    img -= np.min(img)
    img /= np.max(img)
    img *= 255
    return img
```

Trasformasi Logaritma Konstanta 1

```python
img_log_transformed = log_transformation(img, 1)
```

```python
hstacked = np.hstack((img, img_log_transformed))
```

Menampilkan Gambar

```python
plt.imshow(hstacked, cmap="gray")
plt.show()
plt.imshow(img_log_transformed, cmap="gray")
plt.show()
```

Transformasi Pada Array

```python
a = np.array([5, 50, 150, 254, 255], dtype=np.uint8)
a_log = log_transformation(a, 1)
a_log = a_log - np.min(a_log)
a_log = a_log / np.max(a_log)
a_log = a_log * 255

print("Original values:", a)
print("Log-transformed values:", a_log)
```

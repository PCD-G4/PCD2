# Transofrmasi Negatif

Import Libraries

```python
import gdown
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

Download Gambar

```python
url = "https://drive.google.com/uc?id=1jwrW5d4AmUutLl0hUscOePqimXO8IBFS"
output = "cat.jpg"
gdown.download(url, output, quiet=False)
```

Gambar RGB

```python
image_rgb = Image.open(output)
image_rgb = image_rgb.convert('RGB')
image_rgb
```

Transformasi Negatif

```python
image_np = np.array(image_rgb)
image_negatif = 255 - image_np

plt.figure(figsize=(5, 5))
plt.imshow(image_negatif)
plt.title("Gambar Negatif")
plt.axis("off")
plt.show()
```

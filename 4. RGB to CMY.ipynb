# RGB to CMY

Import Libraries

```python
import gdown
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

Download Gambar

```python
url = "https://drive.google.com/uc?id=1jwrW5d4AmUutLl0hUscOePqimXO8IBFS"
output = "cat.jpg"
gdown.download(url, output, quiet=False)
```

Gambar Awal RGB

```python
image_rgb = Image.open(output)
image_rgb = image_rgb.convert('RGB')
image_rgb
```

Convert Gambar RGB to CMY

```python
image_np = np.array(image_rgb) / 255.0
C, M, Y = [(1 - image_np[:, :, i]) * 255 for i in range(3)]

plt.figure(figsize=(15, 5))
for i, (channel, title) in enumerate(zip([C, M, Y], ['Cyan', 'Magenta', 'Yellow'])):
    plt.subplot(1, 3, i + 1)
    plt.imshow(channel.astype(np.uint8), cmap='gray')
    plt.title(title)
    plt.axis('off')
plt.show()
```

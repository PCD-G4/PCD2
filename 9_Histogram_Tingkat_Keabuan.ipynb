```python
import cv2
import matplotlib.pyplot as plt
import math
import numpy as np
image = cv2.imread('output_gamma_3.2.jpg')
plt.imshow(image)
plt.show()
```

```python
img_height = image.shape[0]
img_width = image.shape[1]
img_channel = image.shape[2]
```

```python
hg = np.zeros((256))
```

```python
for x in range(0, 256):
    hg[x] = 0
```

```python
for y in range(0, img_height):
    for x in range(0, img_width):
        gray = image[y][x][0]
        hg[gray] += 1
```

```python
bins = np.linspace(0, 256, 100)
plt.hist(hg, bins, color="black", alpha=0.5)
plt.title("Histogram")
plt.show()
```

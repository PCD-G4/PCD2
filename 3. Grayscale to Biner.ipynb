# Grayscale to Biner

Import Libraries

```python
from PIL import Image
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import gdown
```

Download Gambar

```python
url = "https://drive.google.com/uc?id=1jwrW5d4AmUutLl0hUscOePqimXO8IBFS"
output = "cat.jpg"
gdown.download(url, output, quiet=False)
```

Gambar Awal Grayscale

```python
image_grayscale = Image.open(output).convert('L')
image_grayscale
```

Konversi Gambar menjadi Biner

```python
image_binary = image_grayscale.point(lambda p: 255 if p > 128 else 0)
image_binary
```

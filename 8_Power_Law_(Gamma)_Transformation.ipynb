# Power Law (Gamma) Transformation
Import Libraries

```python
import cv2
import matplotlib.pyplot as plt
import math
import numpy as np
```

Read Image

```python
image = cv2.imread('cat_pic.jpg')
plt.imshow(image)
plt.show()
```

Plotting Histogram Gambar

```python
color = ('b', 'g', 'r')

for i, col in enumerate(color):

    histr = cv2.calcHist([image],
                         [i], None,
                         [256],
                         [0, 256])

    plt.plot(histr, color = col)

    # Limit X - axis to 256
    plt.xlim([0, 256])

plt.show()
```

Koreksi Gamma dengan 4 percobaan nilai

```python
for gamma in [0.1, 0.5, 1.2, 2.2, 3.2]:

    gamma_corrected = np.array(255*(image / 255) ** gamma, dtype = 'uint8')
    plt.imshow(gamma_corrected)
    plt.show()
```

```python
color = ('b', 'g', 'r')

for i, col in enumerate(color):

    histr = cv2.calcHist([gamma_corrected],
                         [i], None,
                         [256],
                         [0, 256])

    plt.plot(histr, color = col)
    plt.xlim([0, 256])

plt.show()
```

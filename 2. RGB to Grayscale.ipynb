# RGB to Grayscale

Import Libraries

```python
from PIL import Image
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import gdown
```

Download Gambar

```python
url = "https://drive.google.com/uc?id=1jwrW5d4AmUutLl0hUscOePqimXO8IBFS"
output = "cat.jpg"
gdown.download(url, output, quiet=False)
```

Gambar Awal RGB

```python
image_rgb = Image.open(output)
image_rgb = image_rgb.convert('RGB')
image_rgb
```

Convert Gambar menjadi Grayscale

```python
image_grayscale = image_rgb.convert('L')
image_grayscale
```
